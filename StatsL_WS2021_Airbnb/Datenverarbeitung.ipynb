# Calendar Datensatz

```python
# Calendar Datei einlesen
df_calendar = pd.read_csv(r"C:\Users\leand\Desktop\Seminar\calendar.csv")
# Daten anschauen
df_calendar.head(5)
```

```python
# Anzahl Spalten und Reihen
print("Reihen:" , df_calendar.shape[0])
print("Spalten:" , df_calendar.shape[1])

# Fehlende Werte
df_calendar.isnull().sum()
```

```python
# Fehlende Werte löschen
# price - adjusted_price:
df_calendar.dropna(subset = ["price"], inplace=True)
# minimum_nights - maximum_nights::
df_calendar.dropna(subset = ["minimum_nights"], inplace=True)
```

```python
# Überprüfung
df_calendar.isnull().sum()
```

```python
# Dubletten Check
print(df_calendar.duplicated(subset=None, keep='first'))
print("Summe der Duplikate:" , df_calendar.duplicated(subset=None, keep='first').sum())
```

```python
# Price Daten zu float umwandeln
df_calendar['price'] = df_calendar.price.replace('[\$,]', '', regex=True).astype(float)
df_calendar['adjusted_price'] = df_calendar.adjusted_price.replace('[\$,]', '', regex=True).astype(float)
```

```python
# Neue Spalten einfügen (year, month, day)
df_calendar['year'] = pd.DatetimeIndex(df_calendar.date).year
df_calendar['month'] = pd.DatetimeIndex(df_calendar.date).month
df_calendar['day'] = pd.DatetimeIndex(df_calendar.date).day
```

```python
# Daten mit (year, month, day) anschauen
df_calendar.head(5)
```

```python
# Es gibt in df_calendar Daten für 01.10.2022
df_calendar[df_calendar.isin(["2022-10-01"]).any(axis=1)]
```

```python
# und für 02.10.2022. Das soll so nicht sein
df_calendar[df_calendar.isin(["2022-10-02"]).any(axis=1)]
```

```python
# Diese Daten sollen gelöscht werden
indexOne = df_calendar[df_calendar['date'] == '2022-10-01'].index
indexTwo = df_calendar[df_calendar['date'] == '2022-10-02'].index
df_calendar.drop(indexOne, inplace = True)
df_calendar.drop(indexTwo, inplace = True)
```
